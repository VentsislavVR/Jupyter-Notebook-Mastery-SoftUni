```python
import numpy as np
# evenly sampled time at 200ms intervals
t = np.arange(0.,5.,0.2)
# red dashes , blue squares and green triangles
plt.plot(t,t, 'r--',t,t**2 , 'bs',t, t**3 , 'g^')
plt.show()
```

```python
# plotimg labaled data 

import pandas as pd 
data = pd.DataFrame({
    'a':np.arange(50),
    'c': np.random.randint(0,50,50),
    'd': np.random.randn(50)
    })
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b' , c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

```python
names = ['group_a','group_b','group_c']
values = [1, 10, 100]

plt.figure(figsize=(9,3))

plt.subplot(131)
plt.bar(names,values)

plt.subplot(132)
plt.scatter(names,values)

plt.subplot(133)
plt.plot(names,values)

plt.suptitle('Categorical Plotting')
plt.show()
```

```python
# toy data 
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin( x**2 )
```

```python
plt.plot(x,y,linewidth=10.0)
plt.show()
```

```python
# name yur axis
# plt.rcParams['font.size'] = '10'

profit = np.array([23,29,34])
year = np.array([2021,2022,2023]).astype('str')

plt.plot(year,profit)
plt.xlabel('Period',fontsize = 15)
plt.ylabel('Profit',fontsize = 25)
plt.title('Profit by year',fontsize = 10)
plt.show()
```

```python
fig,ax = plt.subplots()
ax.plot(x,y)
ax.set_title('a single plot')
plt.show()
```

```python
fig,axs = plt.subplots(2)
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x,y)
axs[1].plot(x,-y)
plt.show()
```

```python
fig,(ax1,ax2) = plt.subplots(1,2) # with unpacking
fig.suptitle('Horizontally stacked subplots')
ax1.plot(x,y)
ax2.plot(x,-y)
plt.show()
```

```python
fig,axs = plt.subplots(2,2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y,'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')
for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')
for ax in axs.flat:
    ax.label_outer()
plt.show()
```

```python
def example_plot(ax):
    ax.plot([1 ,2])
    ax.set_xlabel('x-label',fontsize=12)
    ax.set_ylabel('y-label',fontsize=12)
    ax.set_title('Title',fontsize=15)
```

```python
fix, axs = plt.subplots(nrows=2,ncols=2,layout='tight')
for ax in axs.flat:
    example_plot(ax)
```
